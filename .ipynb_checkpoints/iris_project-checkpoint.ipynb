```python
def euclidean_dist(a,b):
    if len(a) != len(b):
        return None

    return np.sqrt(np.sum((a - b)**2))

def KNN(input_col, species , point , k):


    distance = []
    for i in range(0 , len(input_col)):
        
        dist = euclidean_dist(np.array(input_col[i]) , np.array(point))
        distance.append((dist , species[i]))
        
    distance.sort(key = lambda x : x[0])
    k_nearest = [j for _,j in distance[:k]]

    return Counter(k_nearest).most_common(1)[0][0]



data = df.copy()

input_data_4d = []
input_spe = []
for i , row in data.iterrows():
    input_data_4d.append([row['sepal_length'] , row['sepal_width'] , row['petal_length'] , row['petal_width']])
    input_spe.append(row['species'])
    
k = 5
point = [5.00600,	3.418000,	1.464000,	0.24400]

KNN(input_data_4d , input_spe , point , k)
```

```python
def split_dataset(data , spe , per):
    test_inp , test_out , train_inp , train_out = [] , [] , [] , []
    per_20 = []

    number_test  = int((len(data) * per)/100)
    indices  = np.random.choice(len(data) , size = number_test , replace= False)

    for idx , j in enumerate(data):

        if idx in indices:
            train_inp.append(j)
            train_out.append(spe[idx])
        else:
            test_inp.append(j)
            test_out.append(spe[idx])

    return test_inp , test_out , train_inp , train_out
    

data = df.copy()
input_data_4d = []
input_spe = []
for i , row in data.iterrows():
    input_data_4d.append([row['sepal_length'] , row['sepal_width'] , row['petal_length'] , row['petal_width']])
    input_spe.append(row['species'])

per = 30
test_inp , test_out , train_inp , train_out = split_dataset(input_data_4d , input_spe , per)

len(test_inp) , len(test_out) , len(train_inp) , len(train_out)
```

```python
def accuracy(test_inp , test_out , train_inp , train_out , k):

    count = 0
    for i in range(len(train_inp)):
        if train_out[i] == KNN(test_inp , test_out ,train_inp[i] , k):
            count += 1

    return (count / len(train_inp)) * 100 


k = 5
acc = accuracy(test_inp , test_out , train_inp , train_out , k)
acc
```

```python
k_value = []
acc1 = []
index = []
count = 0
for i in range(0,30):
    test_inp , test_out , train_inp , train_out = split_dataset(input_data_4d , input_spe , 30)
    for k in range(1,16):
        acc = accuracy(test_inp , test_out , train_inp , train_out , k)
        k_value.append(k)
        acc1.append(acc)
        index.append(count)
        count +=1 
 
```
