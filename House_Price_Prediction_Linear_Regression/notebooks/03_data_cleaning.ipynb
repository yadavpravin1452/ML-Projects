```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('HousePricePrediction.xlsx')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(columns = ['Id'] , inplace = True)
```

```python
numeric_df = df.select_dtypes(include = ['int64' , 'float64'])
numeric_df.shape
```

Categorical features are excluded for the baseline numeric only model

```python
numeric_df = numeric_df[numeric_df['SalePrice'].notnull()]
numeric_df['SalePrice'] = numeric_df['SalePrice'].fillna(numeric_df['SalePrice'].mean())
numeric_df['TotalBsmtSF'] = numeric_df['TotalBsmtSF'].fillna(numeric_df['TotalBsmtSF'].mean())
```

Rows with missing SalePrice are removed since the target Variable cannot be imputed.
These rows represent test only Samples and are excluded from model training.

```python
numeric_df.isnull().sum()
```

```python
numeric_df.columns
```

```python
selected_feature = ['TotalBsmtSF' , 'YearBuilt' , 'YearRemodAdd' , 'LotArea' , 'OverallCond']

final_df = numeric_df[selected_feature + ['SalePrice']]
```

Features selected based on correlation analysis with SalePrice.
